# Tabela com Informações Gerais

```python
all_teams = list(set([*games_df['Mandante'].unique(), *games_df['Visitante'].unique()]))
all_teams
```

```python
times_stats = []
for team in all_teams:
    vitorias = games_df[games_df['Vencedor'] == team].shape[0]
    conquistas = games_df[(games_df['Vencedor'] == team) & (games_df['Defensor do Título'] != team)].shape[0]
    defesas = games_df[(games_df['Vencedor'] == team) & (games_df['Defensor do Título'] == team)].shape[0]
    derrotas = games_df[ ((games_df['Mandante'] == team) | (games_df['Visitante'] == team)) & (games_df['Vencedor'] != team) ].shape[0]
    
    jogos = vitorias + derrotas
    
    assert jogos == games_df[(games_df['Mandante'] == team) | (games_df['Visitante'] == team)].shape[0]
    assert conquistas + defesas == vitorias

    pontos_feitos = games_df[(games_df['Mandante'] == team)]['Pontos Mandante'].sum()
    pontos_feitos += games_df[games_df['Visitante'] == team]['Pontos Visitante'].sum()

    pontos_sofridos = games_df[(games_df['Mandante'] == team)]['Pontos Visitante'].sum()
    pontos_sofridos += games_df[games_df['Visitante'] == team]['Pontos Mandante'].sum()

    saldo_pontos = pontos_feitos - pontos_sofridos

    times_stats.append({'Time': team,
                        'Jogos': jogos,
                        'Vitórias': vitorias, 
                        'Conquistas': conquistas, 
                        'Defesas': defesas,
                        'Derrotas': derrotas,
                        'Pontos Feitos': pontos_feitos, 
                        'Pontos Sofridos': pontos_sofridos, 
                        'Saldo de Pontos': saldo_pontos})


general_info = pd.DataFrame(times_stats)
```

```python
general_info = general_info.sort_values(by=['Vitórias'], ascending=False)
```

## Utils

```python
def tournament_postprocess(tournament):
    try:
        split_tournament = tournament.split('-')
    except Exception:
        return tournament
    
    split_tournament = ['BFA' if word == 'bfa' else word.capitalize() for word in split_tournament]
    new_tournament = ' '.join(split_tournament[:-1])

    return new_tournament
    
```

```python
torneios = games_df['Torneio']
```

```python
torneios = torneios.apply(lambda x: tournament_postprocess(x))
```

```python
torneios = torneios.value_counts().sort_values(ascending=False)
torneios.head(10)
```

# Confrontos mais Realizados

```python
confrontos = games_df.apply(lambda x: (x['Mandante'], x['Visitante']), axis=1)
len(confrontos)
```

```python
confrontos = games_df.apply(lambda x: tuple(sorted((x['Mandante'], x['Visitante']))), axis=1)
confrontos = list(set(confrontos))
len(confrontos)
```

```python
confrontos
```

```python
confrontos_list = []
for confronto in confrontos:
    time_1 = confronto[0]
    time_2 = confronto[1]

    qtd_confrontos = games_df[((games_df['Mandante'] == time_1) & (games_df['Visitante'] == time_2)) | ((games_df['Mandante'] == time_2) & (games_df['Visitante'] == time_1))].shape[0]
    confrontos_list.append({
        "Confronto": f"{time_1} x {time_2}",
        "Quantidade de Jogos": qtd_confrontos 
    })

confrontos_df = pd.DataFrame(confrontos_list)
```

```python
confrontos_df = confrontos_df.sort_values(by=['Quantidade de Jogos'], ascending=False)
```

# Venceu primeira partida pelo cinturão do ano

```python
anos = list(games_df['Data'].dt.year.astype(int).unique())
anos
```

```python
primeiro_campeao_ano_list = []
for ano in anos:
    campeao = games_df[games_df['Data'].dt.year == ano].iloc[0]['Vencedor']

    primeiro_campeao_ano_list.append({
        'Ano': ano,
        'Time': campeao
    })

primeiro_campeao_ano_df = pd.DataFrame(primeiro_campeao_ano_list)
```

```python
primeiro_campeao_ano_df = primeiro_campeao_ano_df.sort_values(by=['Ano'], ascending=True)
primeiro_campeao_ano_df
```

# Venceu última partida pelo cinturão

```python
ultimo_campeao_ano_list = []
for ano in anos:
    campeao = games_df[games_df['Data'].dt.year == ano].iloc[-1]['Vencedor']

    ultimo_campeao_ano_list.append({
        'Ano': ano,
        'Time': campeao
    })

ultimo_campeao_ano_df = pd.DataFrame(ultimo_campeao_ano_list)
```

# Primeiras Vezes

```python
times_primeiras_vezes_list = []
for team in all_teams:
    vitorias_df = games_df[games_df['Vencedor'] == team]

    if vitorias_df.shape[0] == 0:
        continue

    times_primeiras_vezes_list.append({'Time': team,
                                       'Data': vitorias_df.iloc[0]['Data'].date()})


primeiras_vezes_df = pd.DataFrame(times_primeiras_vezes_list)
```

```python
primeiras_vezes_df = primeiras_vezes_df.sort_values(by=['Data'], ascending=True)
primeiras_vezes_df
```
