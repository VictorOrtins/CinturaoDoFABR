# First Algorithm Idea

```python
first_game = games_df.iloc[0]
first_champion = first_game['Vencedor']
first_game_date = first_game['Data']

cinturao_games = pd.DataFrame([first_game], index=None)
```

```python
cinturao_games.head()
```

```python
cinturao_games = pd.DataFrame([first_game], index=None)
current_game = first_game
current_champion = first_champion
current_game_date = first_game_date

while True:
    try:
        current_game = games_df[ ((games_df['Mandante'] == current_champion) | (games_df['Visitante'] == current_champion)) & (games_df['Data'] > current_game_date) ].iloc[0]
    except IndexError:
        break
    
    current_champion = current_game['Vencedor']
    current_game_date = current_game['Data']

    cinturao_games = pd.concat([cinturao_games, current_game.to_frame().T], ignore_index=True)
```
